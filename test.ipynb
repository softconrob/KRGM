```python
import pandas as pd

df = pd.read_csv('df_fifa.csv')
df.head()
```

```python
# map the player position to a number, based on what position they play
# the position is a string, containing various positions separated by a comma
# we want to convert the string to a list
# so split the string, then map the position string to a number
h = df['player_positions'].apply(lambda x: x.split(','))

# create a list of all the positions
positions = []
for i in h:
    for j in i:
        positions.append(j.strip())

# remove duplicates
positions = list(set(positions))

# create a dictionary of positions
pos_dict = {}
for i, j in enumerate(positions):
    pos_dict[j] = i
```
